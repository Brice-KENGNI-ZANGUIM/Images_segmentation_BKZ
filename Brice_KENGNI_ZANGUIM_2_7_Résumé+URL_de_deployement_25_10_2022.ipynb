# <strong> B -  Chargement des données

```python
df1 = pd.read_csv("recap_Modèle_Simple_sur_mesure.csv")
```

```python
df_linknet = pd.read_csv("recap_Transfert_learning_LinkNET.csv").drop(['Unnamed: 0'], axis = 1 )
```

```python
df_pspnet = pd.read_csv("recap_Transfert_learning_PSPNET.csv").drop(['Unnamed: 0'], axis = 1 )
```

```python
df4 = pd.read_csv("recap_Transfert_learning_UNET.csv")
```

```python
df5 = pd.read_csv("recap_Transfert_learning_UNET+augmentation.csv")
```

```python
df6 = pd.read_csv("recap_Transfert_learning_UNET+512-32.csv")
```

```python
df7 = pd.read_csv("recap_Transfert_learning_UNET+512-32+sigmoid.csv")
```

```python
df8 = pd.read_csv("recap_UNET.csv")
```

# <strong> C - UNET

## <strong> C.1 - UNET avec augmentation

```python
df_unet_aug = df5.drop(['Unnamed: 0'], axis = 1 )
df_unet_aug.style.highlight_min(axis=0, color='red').highlight_max(axis=0, color='green')
```

## <strong> C.1 - UNET sans augmentation : filtre 256 - 128 - 64 - 32 - 16

```python
df_unet_0 = pd.concat([df4, df8], ignore_index=True).drop(['Unnamed: 0'], axis = 1 )
```

```python
df_unet_0.style.highlight_min(axis=0, color='red').highlight_max(axis=0, color='green')
```

## <strong> C.1 - UNET sans augmentation : filtre 512 - 256 - 128 - 64 - 32

```python
df_unet_1 = pd.concat([ df6, df7], ignore_index=True).drop(['Unnamed: 0'], axis = 1 )
```

```python
df_unet_1.style.highlight_min(axis=0, color='red').highlight_max(axis=0, color='green')
```

## <strong> C.1 - UNET Résumé

```python
df_unet = pd.concat([ df_unet_aug, df_unet_0, df_unet_1], ignore_index=True)
```

```python
df_unet.style.highlight_min(axis=0, color='red').highlight_max(axis=0, color='green')
```

```python
df_unet.groupby(["modèle","Loss","Augmentation"] ).sum().style.highlight_min(axis=0, color='red').highlight_max(axis=0, color='green') 
```
